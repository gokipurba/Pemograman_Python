```
#Perthitungan PathLoss
def PL (d,f,x,htx,hrx):

    #Persamaan Nilai a(hr)
    ahr=((1.1*np.log10(f))-0.7)*hrx-(1.56*np.log10(f)-0.8)

    #persamaan Nilai CM
    if x == "UBN" or "ubn":
      CM = 0
    elif x=="SUBN" or "subn":
      CM = -2*((np.log10*(f/28))**2) - 5.4
    elif x=="OPN" or "opn":
      CM = -4.78*((np.log10(f))**2)+18.33*np.log10(f)-40.94

    #C1 dan C2 
    a = f >=400
    b = f <=1500
    c = f >1500
    d = f <=2000
    if a and b :
      C1 = 69.55
      C2 = 26.16
    elif c and d:
      C1 = 46.3
      C2 = 33.9
    else:
      C1 = 0
      C2 = 0    
    #COST 231 Pathloss Model
    lp = C1+(C2*np.log10(f))-(13.83*np.log10(htx))-ahr+((44.9-6.55*np.log10(htx))*np.log10(d))+CM
    return lp 

```
